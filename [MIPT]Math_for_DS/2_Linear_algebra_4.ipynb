### 1. Вычислите расстояние от точки (1, 1) до прямой,  проходящей через точки (−2, 26) и (10, −9).

Найдем уравнение прямой по 2 точкам, используя канонический вид уравнения прямой

$\dfrac{x-(-2)}{10-(-2)}=\dfrac{y-26}{-9-26}$  ⇒  $(x+2)/12=-(y-26)/35$  ⇒  $35(x+2)=-12(y-26)$  ⇒  $35x+12y+70-12*26=0$   ⇒ 

$35x+12y-242=0$ 

Воспользуемся формулой для нахождения расстояния до прямой  $\dfrac{|Ax_1+By_1+C|}{\sqrt{A^2+B^2}}$

$\dfrac{|35+12-242|}{\sqrt{35^2+12^2}}=\dfrac{195}{\sqrt{1225+144}}=\dfrac{195}{37}$

Ответ: 195/37

### 8. Докажите, что для любого прямоугольника ABCD и произвольной
точки пространства M имеют место равенства

$(\vec{MA}, \vec{MC})=(\vec{MB},\vec{MD})$

$|\vec{MA}|^2+|\vec{MC}|^2=|\vec{MB}|^2+|\vec{MD}|^2$

Для наглядности можно нарисовать координатную плоскость, изобразить прямоугольник с координатами A(x0, yk), B(xk, yk), C(xk, y0), D(x0, y0) и точку M(xm, ym)

Так как мы не привязываемся к конкретным значениям координат, то можем считать что дальнейшие рассуждения будут выполняться для произвольных координат

Определим вектора, указанные в равенства с помощью координат точек

$\vec{MA}=(x_0-x_m, y_k-y_m)$

$\vec{MC}=(x_k-x_m, y_0-y_m)$

$\vec{MB}=(x_k-x_m, y_k-y_m)$

$\vec{MD}=(x_0-x_m, y_0-y_m)$

Получим скалярное произведения векторов

$(\vec{MA}, \vec{MC})=(x_0-x_m)(x_k-x_m)+(y_k-y_m)(y_0-y_m)$

$(\vec{MB},\vec{MD})=(x_k-x_m)(x_0-x_m)+(y_0-y_m)(y_k-y_m)$

Значения выражений будут одинаковыми, так отличие только в порядке сомножителей

По этой же причине, если мы получим длины векторов, а затем их возведем в квадрат, то также будет наблюдаться верное равенство
