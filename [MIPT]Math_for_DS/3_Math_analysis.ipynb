### 1. **Корни квадратного уравнения**

Один из корней уравнения x^2 + 2x + a = 0 равен 7. Найдите второй корень.

По теореме Виета:

$x_1+x_2=-b/a$

В нашем случае: 

$7+x=-2$

Из чего следует, что x=-9

Ответ: -9

### 2. **Кубическое уравнение**

Решите уравнение $2x^3−7x^2+9=0$.

В ответе запишите корни через запятую с пробелом в возрастающем порядке

Можно заметить, что x=-1 является корнем уравнения

По теорему Безу можем представить многочлен как P(x)=(x+1)Q(x)=0

$Q(x)=\dfrac{2x^3−7x^2+9}{x+1}=2x^2-9x+9$

Решим квадратное уравнение через дескриминант

$D=\sqrt{b^2-4ac}=\sqrt{81-4 * 2 * 9}=\sqrt{9}=3$

$x_2=\dfrac{9-3}{4}=6/4=1.5$

$x_3=\dfrac{9+3}{4}=3$

Ответ: -1, 1.5, 3

### 3. **Вершина параболы**

Найдите координаты вершины параболы $y=2x^2−5x+3$.

Ответ запишите в формате (x, y). Например, "(-1.2, 0.25)"

Вершина параболы находится на середине отрезка между точками пересечения параболы оси абсцисс.

$x_в=\dfrac{x_1+x_2}{2}=\dfrac{-b-b}{4a}=-\dfrac{b}{2a}=\dfrac{5}{4}=1.25$

$y_в=2x_в^2-5x_в+3=3.125-12.5+3=-0.125$

Ответ: (1.25, -0.125)

### 4. **Синус 36 градусов**

Известно, что sin(π/10)≈0.31, а cos(π/10)≈0.95. Вычислите sin(π/5)

Ответ округлите до двух десятичных цифр. Пример: "0.14"

Используя формулу из тригонометрии $sin(2x)=2sin(x)cos(x)$

$sin(\pi/5)=sin(2\pi/10)=2sin(\pi/10)cos(\pi/10)=2 * 0.31 * 0.95=0.589$

Ответ: 0.59
