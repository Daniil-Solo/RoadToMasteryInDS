## Множества 

__Задача:__

Из 100 студентов английский язык изучают 28 человек, немецкий – 30 человек, французский – 42 человека, английский и немецкий – 8 человек, английский и французский – 10, немецкий и
французский – 5. Все три языка изучают три студента. Сколько студентов изучают только один язык? Сколько студентов не изучают ни одного языка?

__Решение:__

A - английский, D - немецкий, F - французский

$|A|=28, |D|=30, |F|=42, |O|=?$

$|A\cap D|=8, |A\cap F|=10, |D\cap F|=5$

$|A\cap D \cap F|=3$

$|A\cup D \cup F \cup O|=100$

__Вывод:__

Таким образом, только английский выбрали 13 человек, только немецкий 20 человек, только французский 30 человек. А 20 человек не выбрали ни одного предмета

## Булева алгебра

__Задача:__

Упростите следующие выражения:
1. y ∨ x ∧ ¬x ∨ x ∧ (y ∨ y)
2. (a ⇒ b) ⇒ (a ∧ c ⇒ b ∧ c)

Докажите, что следующие выражения верны:

3. a ⇒ (b ⇒ a ∧ b)
4. (a ∧ b ⇒ c) ⇒ (a ⇒ (b ⇒ c))

__Решение:__

1. y ∨ x ∧ ¬x ∨ x ∧ (y ∨ y) = y ∨ 0 ∨ x ∧ y  = y ∨ x ∧ y = y
2. (a ⇒ b) ⇒ (a ∧ c ⇒ b ∧ c) = (¬a ∨ b) ⇒ (¬(a ∧ c) ∨ (b ∧ c)) = ¬(¬a ∨ b) ∨ (¬a ∨ ¬c ∨ (b ∧ c)) = (a ∧ ¬b) ∨ ¬(a ∧ c) ∨ (b ∧ c) = 1 (по таблице истинности)
3. a ⇒ (b ⇒ a ∧ b) = ¬a ∨ ¬b ∨ a ∧ b = ¬(a ∧ b) ∨  (a ∧ b) = 1
4. (a ∧ b ⇒ c) ⇒ (a ⇒ (b ⇒ c)) = ¬(¬(a ∧ b) ∨ c) ∨ (¬a ∨ (¬b ∨ c)) = ¬(¬a ∨ ¬b ∨ c) ∨ (¬a ∨ ¬b ∨ c) = 1
